```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame['label']= breast_cancer_dataset.target
```

```python
X= data_frame.drop(columns='label',axis=1)
Y= data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# STANDARDISE THE DATA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
# BUILDING THE NEURAL NETWORK
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model =keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Convert to NumPy arrays if they are Pandas objects
if isinstance(X_train, pd.DataFrame):
    X_train = X_train.to_numpy()

if isinstance(Y_train, pd.Series):
    Y_train = Y_train.to_numpy()

# Fit the model
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)

#history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# VISUALIZING ACCURACY AND LOSS
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")

plt.legend(['training data', 'validation data'], loc='center right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model accuracy")
plt.xlabel("epoch")
plt.ylabel("loss")

plt.legend(['training data', 'validation data'], loc='center right')
```

```python
# Check if they are numpy arrays and print their shapes
print("X_test_std type:", type(X_test_std))
print("Y_test type:", type(Y_test))
print("X_test_std shape:", X_test_std.shape)
print("Y_test shape:", Y_test.shape)
```

```python
try:
    loss, accuracy = model.evaluate(X_test_std, Y_test)
    print("Loss:", loss)
    print("Accuracy:", accuracy)
except Exception as e:
    print("Error during evaluation:", e)
```
